```python
# DSA210 - 3D Printing Project
# Data Cleaning, EDA, and Initial Hypothesis Tests

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# ----------------------------
# 1. Load dataset
# ----------------------------
df = pd.read_csv("data.csv")

print("First 5 rows:")
display(df.head())

print("\nShape of dataset:")
print(df.shape)

print("\nColumn names:")
print(df.columns)

print("\nDataset info:")
print(df.info())

print("\nDescriptive statistics:")
display(df.describe())

print("\nMissing values in each column:")
print(df.isnull().sum())

print("\nNumber of duplicated rows:")
print(df.duplicated().sum())

# ----------------------------
# 2. Rename incorrect column
# ----------------------------
df = df.rename(columns={"tension_strenght": "tensile_strength"})

print("\nUpdated column names:")
print(df.columns)

# ----------------------------
# 3. Remove duplicates if any
# ----------------------------
df = df.drop_duplicates()

# ----------------------------
# 4. Save cleaned dataset
# ----------------------------
df.to_csv("cleaned_data.csv", index=False)
print("\nCleaned dataset saved as 'cleaned_data.csv'")

# ----------------------------
# 5. Exploratory Data Analysis
# ----------------------------

# Roughness distribution
plt.figure(figsize=(6,4))
df["roughness"].hist(bins=15)
plt.title("Distribution of Roughness")
plt.xlabel("Roughness")
plt.ylabel("Frequency")
plt.show()

# Tensile strength distribution
plt.figure(figsize=(6,4))
df["tensile_strength"].hist(bins=15)
plt.title("Distribution of Tensile Strength")
plt.xlabel("Tensile Strength")
plt.ylabel("Frequency")
plt.show()

# Elongation distribution
plt.figure(figsize=(6,4))
df["elongation"].hist(bins=15)
plt.title("Distribution of Elongation")
plt.xlabel("Elongation")
plt.ylabel("Frequency")
plt.show()

# Material distribution
plt.figure(figsize=(6,4))
sns.countplot(x="material", data=df)
plt.title("Material Distribution")
plt.xlabel("Material")
plt.ylabel("Count")
plt.show()

# Infill pattern distribution
plt.figure(figsize=(6,4))
sns.countplot(x="infill_pattern", data=df)
plt.title("Infill Pattern Distribution")
plt.xlabel("Infill Pattern")
plt.ylabel("Count")
plt.show()

# Layer height vs roughness
plt.figure(figsize=(6,4))
sns.scatterplot(x="layer_height", y="roughness", data=df)
plt.title("Layer Height vs Roughness")
plt.xlabel("Layer Height")
plt.ylabel("Roughness")
plt.show()

# Infill density vs tensile strength
plt.figure(figsize=(6,4))
sns.scatterplot(x="infill_density", y="tensile_strength", data=df)
plt.title("Infill Density vs Tensile Strength")
plt.xlabel("Infill Density")
plt.ylabel("Tensile Strength")
plt.show()

# Material vs elongation
plt.figure(figsize=(6,4))
sns.boxplot(x="material", y="elongation", data=df)
plt.title("Elongation by Material")
plt.xlabel("Material")
plt.ylabel("Elongation")
plt.show()

# Correlation heatmap
numeric_df = df.select_dtypes(include=np.number)
corr_matrix = numeric_df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# ----------------------------
# 6. Hypothesis Testing
# ----------------------------

print("\nHYPOTHESIS TEST 1")
print("H0: There is no linear relationship between layer height and roughness.")
print("H1: There is a significant positive relationship between layer height and roughness.")

corr1, p_value1 = stats.pearsonr(df["layer_height"], df["roughness"])
print("Correlation:", corr1)
print("p-value:", p_value1)

if p_value1 < 0.05:
    print("Result: Reject H0. There is a statistically significant relationship.")
else:
    print("Result: Fail to reject H0. No statistically significant relationship found.")

print("\nHYPOTHESIS TEST 2")
print("H0: Infill density is not associated with tensile strength.")
print("H1: Higher infill density is associated with higher tensile strength.")

corr2, p_value2 = stats.pearsonr(df["infill_density"], df["tensile_strength"])
print("Correlation:", corr2)
print("p-value:", p_value2)

if p_value2 < 0.05:
    print("Result: Reject H0. There is a statistically significant relationship.")
else:
    print("Result: Fail to reject H0. No statistically significant relationship found.")

print("\nHYPOTHESIS TEST 3")
print("H0: Mean elongation is the same across all material types.")
print("H1: At least one material type has a different mean elongation.")

groups = [group["elongation"].values for name, group in df.groupby("material")]
f_stat, p_value3 = stats.f_oneway(*groups)

print("F-statistic:", f_stat)
print("p-value:", p_value3)

if p_value3 < 0.05:
    print("Result: Reject H0. Material type has a significant effect on elongation.")
else:
    print("Result: Fail to reject H0. No significant difference between material types.")

# ----------------------------
# 7. Short summary
# ----------------------------
print("\nSUMMARY")
print("Data cleaning completed.")
print("EDA visualizations created.")
print("Three initial hypothesis tests were performed.")
print("This notebook covers the 14 April milestone: data collection, EDA, and hypothesis testing.")
```
